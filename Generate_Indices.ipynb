```python
import torch
import numpy as np
import os 

dataset = "openwebtext"
data_dir = os.path.join('data', dataset)
print(data_dir, os.listdir(data_dir))
train_data = np.memmap(os.path.join(data_dir, 'train.bin'), dtype=np.uint16, mode='r')#len = 9035582489
val_data = np.memmap(os.path.join(data_dir, 'val.bin'), dtype=np.uint16, mode='r')  #len = 4434606

train_data.size, val_data.size
```

```python
# Create a generator and set the seed
gen = torch.Generator(device='cpu')
gen.manual_seed(42)
ix_train = torch.randint(train_data.size - 1024, (600000,5*8*12), device = "cpu", generator=gen) 
ix_val = torch.randint(val_data.size - 1024, (600000,5*8*12), device = "cpu", generator=gen)
np.save("data/openwebtext/train_indices.npy",ix_train.numpy())
np.save("data/openwebtext/val_indices.npy",ix_val.numpy())
```

```python
gen = torch.Generator(device='cpu')
gen.manual_seed(42)

ix_train = torch.randint(train_data.size - 1024, (600000//1000,200*12), device = "cpu", generator=gen) 
ix_val = torch.randint(val_data.size - 1024, (600000//1000,200*12), device = "cpu", generator=gen)
np.save("data/openwebtext/eval_train_indices.npy",ix_train.numpy())
np.save("data/openwebtext/eval_val_indices.npy",ix_val.numpy())
```
